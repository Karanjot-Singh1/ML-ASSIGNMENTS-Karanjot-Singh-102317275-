1. (Gaussian Naïve Bayes Classifier) Implement Gaussian Naïve Bayes
Classifier on the Iris dataset from sklearn.datasets using
(i) Step-by-step implementation


```python

import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

d = load_iris()
X = d.data
y = d.target
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.25, random_state=7, stratify=y)

classes = np.unique(ytr)
n, p = Xtr.shape
pri = {}
mu = {}
var = {}
for c in classes:
    Xi = Xtr[ytr == c]
    pri[c] = Xi.shape[0] / n
    mu[c] = Xi.mean(axis=0)
    var[c] = Xi.var(axis=0) + 1e-9  # stability

def gaussian_loglik(x, m, v):
    return -0.5 * np.sum(np.log(2 * np.pi * v)) - 0.5 * np.sum(((x - m) ** 2) / v)

pred = []
for x in Xte:
    scores = []
    for c in classes:
        score = np.log(pri[c]) + gaussian_loglik(x, mu[c], var[c])
        scores.append(score)
    pred.append(np.argmax(scores))
pred = np.array(pred)
print("Stepwise GAUSS NB Accuracy:", round(accuracy_score(yte, pred), 4))
```

(ii) In-built function





```python

from sklearn.naive_bayes import GaussianNB
gn = GaussianNB()
gn.fit(Xtr, ytr)
p = gn.predict(Xte)
print("Sklearn GAUSS NB Accuracy:", round(accuracy_score(yte, p), 4))
```

2. Explore about GridSearchCV toot in scikit-learn. This is a tool that is
often used for tuning hyperparameters of machine learning models. Use
this tool to find the best value of K for K-NN Classifier using any dataset.

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

d = load_iris()
X = d.data
y = d.target
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.25, random_state=9, stratify=y)

kn = KNeighborsClassifier()
grid = {'n_neighbors': list(range(1, 26))}
g = GridSearchCV(kn, grid, cv=5, scoring='accuracy', n_jobs=-1)
g.fit(Xtr, ytr)
best_k = g.best_params_['n_neighbors']
best_cv = g.best_score_
final = KNeighborsClassifier(n_neighbors=best_k).fit(Xtr, ytr)
p = final.predict(Xte)
print("Best K:", best_k, "CV Accuracy:", round(best_cv, 4), "Test Acc:", round(accuracy_score(yte, p), 4))
```
